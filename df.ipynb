```python
from scipy import stats

r, p_value = stats.pearsonr(df['Annual_Income'].fillna(0), df['Num_Credit_Inquiries'].fillna(0), alternative='two-sided')
print("\nHypothesis Test for Correlation between Income and Num_Credit_Inquiries:")
print(f"Correlation coefficient (r): {r}, p-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Do not reject the null hypothesis")
```

```python

df.Monthly_Inhand_Salary = pd.to_numeric(df.Monthly_Inhand_Salary, errors='coerce')
yes_purchase = df[df['Payment_of_Min_Amount'] == 'Yes']['Monthly_Inhand_Salary']
no_purchase = df[df['Payment_of_Min_Amount'] == 'No']['Monthly_Inhand_Salary']
# Before performing Two Independent Samples t-Test, perform Levene test for equal variances.
# The Levene test tests the null hypothesis that all input samples are from populations with equal variances. 
stat, p_value = stats.levene(yes_purchase, no_purchase)
print("\nLevene's test for Having payment of min amount:")
print(f"Test statistic: {stat}, p-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Do not reject the null hypothesis")
# Perform Two Independent Samples t-Test
# This tests whether there is a significant difference in the average purchase amount between female and male ccustomers.
t_stat, p_value = stats.ttest_ind(yes_purchase, no_purchase, equal_var=True, alternative='two-sided')
print("\nTwo Independent Samples t-Test for Female vs. Male Purchase Amount:")
print(f"t-statistic: {t_stat}, p-value: {p_value}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Do not reject the null hypothesis")
```

```python
from scipy import stats

contingency_table = pd.crosstab(df['Occupation'], df['Num_Credit_Card'])
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)
print("\nChi-square (χ2) Test of Independence: Occupation vs. Num_Credit_Card:")
print(f"Chi-square statistic: {chi2_stat}, p-value: {p_value}")
alpha = 0.05
# Interpret the results
if p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Do not reject the null hypothesis")
```
